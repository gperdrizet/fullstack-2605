## 1. Linear algebra with NumPy

```python
import numpy as np
```

### 1.1. Vectors

Ordered list of numbers:

```python
u = np.array([1, 4, 5])
v = np.array([2, -3, 6])

print(f'u: {u}')
print(f'v: {v}')
```

Element-wise addition:

```python
print(f'u + v = {u + v}')
```

```python
print(f'3 + u = {3 + u}')
```

#### 1.1.1 Dot product

Multiply corresponding elements and sum the results:

```python
dot = np.dot(u, v)

print(f'u: {u}')
print(f'v: {v}')
print(f'\nu dot v: {dot}')
```

| u | v | u x v   | product |
|---|---|---------|---------|
| 1 | 2 | 1 x 2   |    2    |
| 4 | -3| 4 x (-3)|   -12   |
| 5 | 6 | 5 x 6   |    30   |

2 + (-12) + 30 = 20

#### 1.1.2. Euclidean norm (magnitude)

'Length' of the vector:

```python
norm_u = np.linalg.norm(u)
print(f'Norm of u: {norm_u:.2f}')
```

### 1.2. Matrices

```python
A = np.array([[4, 7],
              [2, 6]])

B = np.array([[5, 8],
              [3, 4]])

print("A:")
print(A)

print("\nB:")
print(B)
```

#### 1.2.1. Addition & subtraction

```python
print("A:")
print(A)

print("\nB:")
print(B)

print("\nSum:")
print(np.add(A, B))
```

```python
print("A:")
print(A)

print("\nB:")
print(B)

print("\nDifference:")
print(np.subtract(A, B))
```

#### 1.2.2. Scalar multiplication

```python
print("A:")
print(A)

print("\nA * 3")
print(A * 3)
```

#### 1.2.3. Matrix multiplication

Each output value is the dot product of a row in A with a colum from B.

```python
print("A:")
print(A)

print("\nB:")
print(B)

print("\nA @ B:")
print(A @ B)
```

#### 1.2.4. Transpose

Swaps rows and columns in matrix:

```python
print("A:")
print(A)
print()
print("Transpose A:")
print(A.T)
```

#### 1.2.5. Rank

The number of (linearly) independent rows.

```python
rank_A = np.linalg.matrix_rank(A)

print("A:")
print(A)
print()
print(f'Rank of A {rank_A}')
```

```python
C = np.array([[1,2,3],
             [4,5,6],
             [2,4,6]]) # this row is 2 x row 0

rank_C = np.linalg.matrix_rank(C)

print("C:")
print(C)
print()
print(f'Rank of C {rank_C}')
```

## 2. Statistics fundamentals

```python
import statistics
import matplotlib.pyplot as plt
```

```python
scores = np.random.normal(65, 12, 200)
```

### 2.1. Measures of central tendency

Where the center of the the data tends to be.

#### 2.1.1. Mean

```python
# Mean (average) add up observations and divide by
# number of observations
mean = statistics.mean(scores)
print(f'Mean: {round(mean, 2)}')
```

#### 2.1.2. Median

```python
# Median middle number (or half way between the two middle numbers)
median = statistics.median(scores)
print(f'Median: {round(median, 2)}')
```

#### 2.1.3. Mode

```python
brands = ['B', 'A', 'Z', 'A', 'C']

mode = statistics.mode(brands)
print(f'Mode: {mode}')
```

### 2.2. Measure of dispersion (spread)

#### 2.2.1. Range

```python
# Range: highest value - lowest value
data_range = scores.max() - scores.min()
print(f'Data range: {data_range:.2f}')
```

#### 2.2.2. Interquartile range (IQR)

```python
# IQR (interquartile range)
q1 = np.percentile(scores, 25)
q3 = np.percentile(scores, 75)

iqr = q3 - q1

print(f'Q1: {q1:.2f}, Q3: {q3:.2f}')
print(f'IQR: {iqr:.2f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.set_title('Histogram')
ax1.hist(scores, color='black', bins=30)
ax1.axvspan(q1, q3, color='green', alpha=0.5)
ax1.set_xlabel('score')
ax1.set_ylabel('count')

ax2.set_title('Boxplot')
ax2.boxplot(
    scores,
    patch_artist=True,
    boxprops=dict(facecolor='green', alpha=0.4),
)
ax2.set_xlabel('Test')
ax2.set_ylabel('Score')

plt.show()
```

#### 2.2.3. Variance & standard deviation

```python
# Standard deviation is square root of the variance
# variance is the the average of the squared 
# differences from the mean

std = statistics.stdev(scores)
var = statistics.variance(scores)

print(f'Variance: {var}')
print(f'Standard deviation: {std}')
```

```python
plt.title('Standard deviation')
plt.hist(scores, color='black', bins=30)
plt.axvline(mean, color='red', label='mean')
plt.axvspan(
    mean - std, mean + std,
    color='red',
    alpha=0.5,
    label='1 std'
)
plt.axvspan(
    mean - (2 * std), mean + (2 * std),
    color='red',
    alpha=0.1,
    label='2 std'
)
plt.legend()
plt.show()
```

### 3.3. Shape

#### 3.3.1. Skew

```python
right_skew = np.random.beta(2, 7, 1000) * 100
symmetric = np.random.normal(50, 10, 1000)
left_skew = np.random.beta(7, 2, 1000) * 100

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3.5))

ax1.set_title('Skew left')
ax1.hist(left_skew, color='black', bins=30)

ax2.set_title('Symmetric')
ax2.hist(symmetric, color='black', bins=30)

ax3.set_title('Skew right')
ax3.hist(right_skew, color='black', bins=30)

plt.show()
```

#### 3.3.1. Kurtosis

```python
import scipy.stats as sp

x = np.linspace(-5, 5, 100)

plt.title('Kurtosis')
plt.plot(x, sp.uniform.pdf(x, -3, 6), label='kertosis < 0')
plt.plot(x, sp.norm.pdf(x), label='kertosis = 0')
plt.plot(x, sp.laplace.pdf(x), label='kertosis > 0')
plt.legend()
plt.show()
```
