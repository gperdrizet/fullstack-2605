## 4. Array functions & operations

### 4.1. Arithmetic operations

````python
a = np.array([10,20,30,40])
b = np.array([1,2,3,4])
c = np.array([1])
````

````python
print(a + b)
````

### 4.2 Statistical methods

````python
print(np.mean(a))
````

````python
print(np.sum(a))
````

### 4.3. Methods on axes

````python
arr_2d = np.array([[
    [10,20,30,40],
    [1,2,3,4],
],[
    [1,2,3,4],
    [3,4,5,6]
]])

print(arr_2d)
````

````python
arr_2d.shape
````

````python
print(np.mean(arr_2d))
````

````python
print(np.mean(arr_2d, axis=0))
````

````python
print(np.mean(arr_2d, axis=1))
````

````python
arr_2d.shape
````

### 4.4. Shape manipulation

````python
arr = np.arange(12)
print(arr)
````

````python
new_arr = arr.reshape(2,6)
print(new_arr)
````

````python
new_arr = arr.reshape(2,7)
print(new_arr)
````
