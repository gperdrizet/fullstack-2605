## 1. Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 1.1. Bar plot

```python
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 15]
```

```python
plt.title('Bar chart')
plt.barh(categories, values, color='grey', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Value')
plt.show()
```

### 1.2. Line plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.title('Line plot')
plt.plot(
    x, y,
    color='red',
    linestyle='--',
    marker='o',
    markerfacecolor='black',
    markeredgecolor='black', 
    markersize=4
)

plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

### 1.3. Scatter plot

```python
x = np.random.rand(100)
y = 2 * x + np.random.rand(100)
```

```python
plt.title('Scatter plot')
plt.scatter(x, y, color='black', s=10)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

### 1.4. Distribution plots

#### 1.4.1. Histograms

```python
data = np.random.normal(0, 1, 10000)
```

```python
plt.title('Histogram')
plt.hist(data, bins=30, color='gray', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

#### 1.4.2. Boxplots

```python
data = [
    np.random.normal(0, 1, 1000),
    np.random.normal(0, 1.5, 1000),
    np.random.normal(0, 2, 1000)
]
```

```python
plt.title('Boxplot')
plt.boxplot(data)
plt.show()
```

### 1.5. Customizing and combining plots
#### 1.5.1. Two series with legend

```python
n = 1000
x = np.random.uniform(0, 10, n)

y1_linear = 0.5 * x + 10
y2_linear = 2.5 * x + 0.1
y3_linear = -1.0 * x + 0.5

y1 = y1_linear + np.random.normal(0, 1, n)
y2 = y2_linear + np.random.normal(0, 1, n)
y3 = y3_linear + np.random.normal(0, 1, n)


plt.title('Two series scatter plot')
plt.scatter(x, y1, label='y1', color='black', s=1)
plt.scatter(x, y2, label='y2', color='red', s=1)
plt.scatter(x, y3, label='y3', color='blue', s=1)
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend(markerscale=5)
plt.show()
```

#### 1.5.2. Multiple plot types, same axies

```python
plt.title('Scatter & line plot')
plt.scatter(x, y1, label='y1', color='black', s=2, alpha=0.5)
plt.scatter(x, y2, label='y2', color='red', s=2, alpha=0.5)

plt.plot(x, y1_linear, color='black')
plt.plot(x, y2_linear, color='red')

plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()
plt.show()
```

#### 1.5.3. Multiple subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs = axs.flatten()

fig.suptitle('Multiple plots example', fontsize=24)

axs[0].set_title('Full data range')
axs[0].scatter(x, y1, color='black')
axs[0].set_xlabel('x value')
axs[0].set_ylabel('y value')

axs[1].set_title('Zoomed in view')
axs[1].scatter(x, y1, color='black')
axs[1].set_xlabel('x value')
axs[1].set_ylabel('y value')
axs[1].set_xlim(6, 8)
axs[1].set_ylim(10, 17)

axs[2].set_title('Data distribution (x)\n n bins = 10')
axs[2].hist(x, color='black')
axs[2].set_xlabel('x value')
axs[2].set_ylabel('Count')

axs[3].set_title('Data distribution (x)\n n bins = 50')
axs[3].hist(x, color='black', bins=50)
axs[3].set_xlabel('x value')
axs[3].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
