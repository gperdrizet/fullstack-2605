## 1. Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 1.1. Bar plot

```python
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 15]
```

```python
plt.title('Bar chart')
plt.barh(categories, values, color='grey', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Value')
plt.show()
```

### 1.2. Line plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.title('Line plot')
plt.plot(
    x, y,
    color='red',
    linestyle='--',
    marker='o',
    markerfacecolor='black',
    markeredgecolor='black', 
    markersize=4
)

plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

### 1.3. Scatter plot

```python
x = np.random.rand(100)
y = 2 * x + np.random.rand(100)
```

```python
plt.title('Scatter plot')
plt.scatter(x, y, color='black', s=10)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

## 1.4. Distribution plots

### 1.4.1. Histograms

```python
data = np.random.normal(0, 1, 10000)
```

```python
plt.title('Histogram')
plt.hist(data, bins=30, color='gray', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

### 1.4.2 Boxplots

```python
data = [
    np.random.normal(0, 1, 1000),
    np.random.normal(0, 1.5, 1000),
    np.random.normal(0, 2, 1000)
]
```

```python
plt.title('Boxplot')
plt.boxplot(data)
plt.show()
```
