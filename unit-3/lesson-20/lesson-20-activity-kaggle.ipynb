2. Add the competition dataset as a notebook source:

   - Click `+ Add Input` and filter by `Competition Datasets`.
   - Select `Playground Series - Season 6, Episode 7`.
3. Update the data-loading cell paths to Kaggle paths:
   - `train_df_path = '/kaggle/input/playground-series-s6e7/train.csv'`
   - `test_df_path = '/kaggle/input/playground-series-s6e7/test.csv'`
4. Update the submission path in section 5 to `submission_file = 'submission.csv'`.
5. Run the notebook and submit from the Kaggle notebook sidebar.
