# Lesson 16: Feature engineering demonstration

This notebook demonstrates key feature engineering techniques for transforming, scaling, and encoding features.

**1. Feature transformations**

- **Log transformation**: Apply logarithmic transformation to reduce skewness in right-skewed data
- **Square root transformation**: Apply square root transformation for moderate skewness reduction
- **Power transformation**: Use Box-Cox or Yeo-Johnson methods to make data more Gaussian-like
- **Quantile transformation**: Map values to uniform or normal distributions using quantiles

**2. Feature scaling**

- **Min-max scaling**: Scale features to a fixed range [0, 1]
- **Standard scaling**: Standardize features to have mean=0 and standard deviation=1

**3. Feature encoding**

- **Ordinal encoding**: Convert ordered categories to integers preserving rank order
- **One-hot encoding**: Create binary columns for each category (nominal data)
- **Feature hashing**: Map high-cardinality categorical features to fixed-size hash space

**4. Grouping operations**

- **Aggregate statistics**: Calculate group-level statistics (mean, median, std, etc.)
- **Group-based features**: Create features based on deviations and ratios from group means

**5. Interaction features**

- **Multiplicative interactions**: Multiply pairs of features to capture joint effects
- **Ratio interactions**: Divide features to capture relative relationships
- **Cross-feature differences**: Subtract features to capture contrasts

**6. Polynomial features**

- **Degree-2 expansion**: Automatically generate all squared terms and pairwise interactions
- **Feature inspection**: Examine the names and count of generated polynomial terms

**7. Dimensionality reduction**

- **PCA**: Use principal component analysis to project high-dimensional features into a lower-dimensional space
- **Explained variance**: Assess how much variance each principal component captures
- **Biplot**: Visualize how original features contribute to the principal components


## Notebook set up

### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import (
    PowerTransformer,
    QuantileTransformer,
    MinMaxScaler,
    StandardScaler,
    OrdinalEncoder,
    OneHotEncoder,
    PolynomialFeatures
)
from sklearn.feature_extraction import FeatureHasher
from sklearn.decomposition import PCA

# Set random seed for reproducibility
np.random.seed(315)
```

### Load data

```python
df = pd.read_csv('https://gperdrizet.github.io/FSA_devops/assets/data/unit2/environmental_data.csv')
```

## 1. Feature transformations

### 1.1. Log transformation

```python
# Select a feature
feature = 'area'

# Apply log transformation (using log1p to handle zeros)
df[f'{feature}_log'] = np.log1p(df[feature])

# Visualize before and after
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

axes[0].set_title('Original distribution')
axes[0].hist(df[feature], bins=50, edgecolor='black', color='grey')
axes[0].set_xlabel(feature)
axes[0].set_ylabel('Frequency')

axes[1].set_title('Log-transformed distribution')
axes[1].hist(df[f'{feature}_log'], bins=50, edgecolor='black', color='grey')
axes[1].set_xlabel(f'{feature}_log')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### 1.2. Square root transformation

```python
# Use a different feature for square root transformation
feature = 'area'

# Apply square root transformation
df[f'{feature}_sqrt'] = np.sqrt(df[feature])

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

axes[0].set_title('Original distribution')
axes[0].hist(df[feature], bins=50, edgecolor='black', color='grey')
axes[0].set_xlabel(feature)
axes[0].set_ylabel('Frequency')

axes[1].set_title('Square root-transformed distribution')
axes[1].hist(df[f'{feature}_sqrt'], bins=50, edgecolor='black', color='grey')
axes[1].set_xlabel(f'{feature}_sqrt')
axes[1].set_ylabel('Frequency')



plt.tight_layout()
plt.show()
```

### 1.3. Power transformation

```python
# Use a feature with outliers for power transformation
feature = 'area'

# Apply power transformation using Yeo-Johnson (handles positive and negative values)
transformer = PowerTransformer(method='yeo-johnson')

# Fit and transform
df[f'{feature}_power'] = transformer.fit_transform(df[[feature]])

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

axes[0].set_title('Original distribution')
axes[0].hist(df[feature], bins=50, edgecolor='black', color='grey')
axes[0].set_xlabel(feature)
axes[0].set_ylabel('Frequency')

axes[1].set_title('Power-transformed distribution')
axes[1].hist(df[f'{feature}_power'], bins=50, edgecolor='black', color='grey')
axes[1].set_xlabel(f'{feature}_power')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### 1.4. Quantile transformation

```python
# Apply quantile transformation to uniform distribution
transformer = QuantileTransformer(output_distribution='uniform')

df[f'{feature}_quantile_uniform'] = transformer.fit_transform(df[[feature]])

# Apply quantile transformation to normal distribution
transformer = QuantileTransformer(output_distribution='normal')

df[f'{feature}_quantile_normal'] = transformer.fit_transform(df[[feature]])

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

axes[0].set_title('Original distribution')
axes[0].hist(df[feature], bins=50, edgecolor='black', color='grey')
axes[0].set_xlabel(feature)
axes[0].set_ylabel('Frequency')

axes[1].set_title('Quantile transform (uniform)')
axes[1].hist(df[f'{feature}_quantile_uniform'], bins=50, edgecolor='black', color='grey')
axes[1].set_xlabel('Uniform [0, 1]')
axes[1].set_ylabel('Frequency')

axes[2].set_xlabel('Normal distribution')
axes[2].set_title('Quantile transform (normal)')
axes[2].hist(df[f'{feature}_quantile_normal'], bins=50, edgecolor='black', color='grey')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## 2. Feature scaling

### 2.1. Min-max scaling

```python
# Select features with different scales
features_to_scale = ['temperature', 'humidity', 'wind_speed']

# Apply Min-Max scaling
scaler = MinMaxScaler()

df_minmax = pd.DataFrame(
    scaler.fit_transform(df[features_to_scale]),
    columns=[f'{col}_minmax' for col in features_to_scale],
    index=df.index
)

# Visualize
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].set_title('Original features (different scales)')
axes[0, 0].boxplot([df[col] for col in features_to_scale], tick_labels=features_to_scale)
axes[0, 0].set_ylabel('Value')

axes[0, 1].set_title('Min-max scaled features')
axes[0, 1].boxplot([df_minmax[col] for col in df_minmax.columns], tick_labels=features_to_scale)
axes[0, 1].set_ylabel('Scaled value [0, 1]')

axes[1, 0].set_title('Temperature distribution (original)')
axes[1, 0].hist(df['temperature'], bins=50, edgecolor='black', color='grey')
axes[1, 0].set_xlabel('Temperature')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].set_title('Temperature distribution (min-max scaled)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].hist(df_minmax['temperature_minmax'], bins=50, edgecolor='black', color='grey')
axes[1, 1].set_xlabel('Scaled temperature [0, 1]')

plt.tight_layout()
plt.show()
```

### 2.2. Standard scaling

```python
# Apply Standard scaling
scaler = StandardScaler()

df_standard = pd.DataFrame(
    scaler.fit_transform(df[features_to_scale]),
    columns=[f'{col}_standard' for col in features_to_scale],
    index=df.index
)

# Visualize
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].set_title('Original features (different scales)')
axes[0, 0].boxplot([df[col] for col in features_to_scale], tick_labels=features_to_scale)
axes[0, 0].set_ylabel('Value')

axes[0, 1].set_title('Standard scaled features')
axes[0, 1].boxplot([df_standard[col] for col in df_standard.columns], tick_labels=features_to_scale)
axes[0, 1].set_ylabel('Standardized value (z-score)')

axes[1, 0].set_title('Temperature distribution (original)')
axes[1, 0].hist(df['temperature'], bins=50, edgecolor='black', color='grey')
axes[1, 0].set_xlabel('Temperature')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].set_title('Temperature distribution (standard scaled)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].hist(df_standard['temperature_standard'], bins=50, edgecolor='black', color='grey')
axes[1, 1].set_xlabel('Standardized temperature (z-score)')

plt.tight_layout()
plt.show()
```

## 3. Feature encoding

### 3.1. Ordinal encoding

```python
# Use the ordinal categorical variable (quality)
# Define the order for ordinal encoding
categories = [['poor', 'fair', 'good', 'excellent']]

# Apply ordinal encoding
encoder = OrdinalEncoder(categories=categories)
df['quality_encoded'] = encoder.fit_transform(df[['quality']])

# Display mapping
print('Ordinal encoding mapping:')
df[['quality', 'quality_encoded']].drop_duplicates().sort_values('quality_encoded')
```

### 3.2. One-hot encoding

```python
# Use the nominal categorical variable (region)
# Apply one-hot encoding
encoder = OneHotEncoder(sparse_output=False, drop='first')
onehot_array = encoder.fit_transform(df[['region']])

# Create dataframe with proper column names
df_onehot = pd.DataFrame(
    onehot_array,
    columns=encoder.get_feature_names_out(['region']),
    index=df.index
)

df_onehot.head(10)
```

### 3.3. Feature hashing

```python
# Use the high-cardinality categorical variable (location_id)
print(f'Number of unique location IDs: {df["location_id"].nunique()}')
print(f'Sample IDs:')
print(df['location_id'].head(10))

# Apply feature hashing
hasher = FeatureHasher(n_features=10, input_type='string')

# Transform to sparse matrix
hashed_features = hasher.transform(df['location_id'].apply(lambda x: [x]))

# Convert to dense array for visualization
hashed_array = hashed_features.toarray()

print(f'\nFeature hashing result:')
print(f'Original feature: {df["location_id"].nunique()} unique categories')
print(f'Hashed features: {hashed_array.shape[1]} fixed dimensions')
print(f'Sparse matrix density: {hashed_features.nnz / (hashed_features.shape[0] * hashed_features.shape[1]):.2%}')

# Show sample of hashed features
hashed_df = pd.DataFrame(
    hashed_array[:5],
    columns=[f'hash_{i}' for i in range(hashed_array.shape[1])]
)
print(f'\nSample hashed features:')
hashed_df.head()
```

## 4. Grouping operations

```python
# Group by region and calculate statistics
grouped_stats = df.groupby('region')['temperature'].agg([
    'mean',
    'median',
    'std',
    'min',
    'max',
    'count'
]).round(2)

print('Temperature statistics by region:')
print(grouped_stats)
```

```python
# Create aggregate features
# Add mean temperature by region
df['temp_mean_by_region'] = df.groupby('region')['temperature'].transform('mean')

# Create deviation from group mean
df['temp_deviation'] = df['temperature'] - df['temp_mean_by_region']

# Create ratio to group mean (add constant to avoid division issues)
df['temp_ratio'] = (df['temperature'] + 50) / (df['temp_mean_by_region'] + 50)

print('New aggregate features:')
print(df[['region', 'temperature', 'temp_mean_by_region', 'temp_deviation', 'temp_ratio']].head(10))
```

## 5. Interaction features

Interaction features capture relationships between two or more variables that a model might not detect from the individual features alone.


### 5.1. Multiplicative and ratio interactions


```python
# Select numerical features for interactions
num_features = ['temperature', 'humidity', 'wind_speed']

# Multiplicative interaction: temperature × humidity (heat index proxy)
df['temp_x_humidity'] = df['temperature'] * df['humidity']

# Ratio interaction: temperature / wind_speed (wind chill proxy, add small constant to avoid division by zero)
df['temp_per_wind'] = df['temperature'] / (df['wind_speed'] + 1e-6)

# Difference interaction: temperature - humidity
df['temp_minus_humidity'] = df['temperature'] - df['humidity']

# Show the new interaction features
print('Sample interaction features:')
print(df[['temperature', 'humidity', 'wind_speed',
          'temp_x_humidity', 'temp_per_wind', 'temp_minus_humidity']].head(10))
```

### 5.2. Visualising interaction features


```python
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

axes[0].set_title('Temperature × humidity')
axes[0].scatter(df['temperature'], df['humidity'],
                c=df['temp_x_humidity'], cmap='viridis', alpha=0.4, s=10)
axes[0].set_xlabel('Temperature')
axes[0].set_ylabel('Humidity')

axes[1].set_title('Temperature / wind speed')
axes[1].scatter(df['temperature'], df['wind_speed'],
                c=df['temp_per_wind'], cmap='plasma', alpha=0.4, s=10)
axes[1].set_xlabel('Temperature')
axes[1].set_ylabel('Wind speed')

axes[2].set_title('Temperature - humidity')
axes[2].hist(df['temp_minus_humidity'], bins=50, edgecolor='black', color='grey')
axes[2].set_xlabel('Temperature - humidity')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## 6. Polynomial features

`PolynomialFeatures` systematically generates all powers up to a given degree and all pairwise (and higher-order) product terms. For degree 2 with features $a$, $b$, $c$ it produces: $a$, $b$, $c$, $a^2$, $ab$, $ac$, $b^2$, $bc$, $c^2$.

### 6.1. Generating degree-2 features


```python
poly_input_features = ['temperature', 'humidity', 'wind_speed']

# Generate degree-2 polynomial features (no bias term needed for feature engineering)
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_array = poly.fit_transform(df[poly_input_features])

poly_feature_names = poly.get_feature_names_out(poly_input_features)
df_poly = pd.DataFrame(poly_array, columns=poly_feature_names, index=df.index)

print(f'Input features:  {len(poly_input_features)}')
print(f'Output features: {len(poly_feature_names)}')
print(f'\nGenerated feature names:')
print(list(poly_feature_names))
```

### 6.2. Comparing original and polynomial features


```python
# Plot original features alongside selected polynomial terms
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
fig.suptitle('Original features vs. degree-2 polynomial expansions', fontsize=11)

# Row 0: original features
for i, col in enumerate(poly_input_features):
    axes[0, i].set_title(f'Original: {col}')
    axes[0, i].hist(df_poly[col], bins=50, edgecolor='black', color='grey')
    axes[0, i].set_ylabel('Frequency')

# Row 1: selected squared / interaction terms
poly_terms = ['temperature^2', 'humidity^2', 'temperature humidity']
for i, col in enumerate(poly_terms):
    axes[1, i].set_title(f'Polynomial: {col}')
    axes[1, i].hist(df_poly[col], bins=50, edgecolor='black', color='steelblue')
    axes[1, i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## 7. Dimensionality reduction

PCA finds the directions of maximum variance in the feature space (principal components) and projects the data onto a lower-dimensional subspace. It is useful for visualisation, noise reduction, and removing multicollinearity before modelling.

### 7.1. Fitting PCA


```python
# Use all numerical features for PCA
pca_input_features = ['temperature', 'humidity', 'wind_speed', 'area']

# Standardise first: PCA is sensitive to scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[pca_input_features])

# Fit PCA keeping all components to inspect variance
pca_full = PCA()
pca_full.fit(X_scaled)

explained_var = pca_full.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

print('Explained variance ratio per component:')
for i, (ev, cv) in enumerate(zip(explained_var, cumulative_var), start=1):
    print(f'  PC{i}: {ev:.4f}  (cumulative: {cv:.4f})')
```

### 7.2. Explained variance plot


```python
n_components = len(pca_input_features)
component_labels = [f'PC{i}' for i in range(1, n_components + 1)]

fig, axes = plt.subplots(1, 2, figsize=(9, 3))

axes[0].set_title('Explained variance per component')
axes[0].bar(component_labels, explained_var, edgecolor='black', color='grey')
axes[0].set_ylabel('Proportion of variance')
axes[0].set_ylim(0, 1)

axes[1].set_title('Cumulative explained variance')
axes[1].plot(component_labels, cumulative_var, marker='o', color='steelblue')
axes[1].axhline(0.95, linestyle='--', color='grey', label='95% threshold')
axes[1].set_ylabel('Cumulative proportion')
axes[1].set_ylim(0, 1.05)
axes[1].legend()

plt.tight_layout()
plt.show()
```

### 7.3. 2D projection and biplot

Project the data onto the first two principal components and overlay the original feature loadings to see which features drive each component.


```python
# Project data to 2 components
pca_2d = PCA(n_components=2)
X_pca = pca_2d.fit_transform(X_scaled)

df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=df.index)
df_pca['region'] = df['region'].values

loadings = pca_2d.components_.T  # shape: (n_features, 2)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Scatter plot coloured by region
regions = df_pca['region'].unique()
colors = plt.cm.tab10(np.linspace(0, 1, len(regions)))

axes[0].set_title('PCA 2D projection (coloured by region)')
for region, color in zip(regions, colors):
    mask = df_pca['region'] == region
    axes[0].scatter(df_pca.loc[mask, 'PC1'], df_pca.loc[mask, 'PC2'],
                    label=region, alpha=0.4, s=10, color=color)
axes[0].set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)')
axes[0].set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)')
axes[0].legend(markerscale=2, fontsize=8)

# Biplot: loadings as arrows
axes[1].set_title('Biplot: feature loadings on PC1 and PC2')
axes[1].scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.2, s=5, color='grey')

scale = np.max(np.abs(X_pca)) * 0.4
for i, feature_name in enumerate(pca_input_features):
    axes[1].annotate(
        '',
        xy=(loadings[i, 0] * scale, loadings[i, 1] * scale),
        xytext=(0, 0),
        arrowprops=dict(arrowstyle='->', color='crimson', lw=1.5)
    )
    axes[1].text(
        loadings[i, 0] * scale * 1.1,
        loadings[i, 1] * scale * 1.1,
        feature_name, color='crimson', fontsize=9, ha='center'
    )

axes[1].set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)')
axes[1].set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)')
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```
